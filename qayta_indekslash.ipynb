```python
from pandas import Series, DataFrame
```

```python
data = {
    'Talaba' : ['Ali', 'Kamron', 'Aziza', 'Nodira', 'Samir', 'Shoxrux', 'Bexruz', 'Bobur'],
    'Matematika':[12.7, 19.8, 25.6, 20.1, 27.2, 19.9, 25.8, 23.4],
    'Informatika':[23.8, 22.5, 18.6, 21.2, 19.3, 21.8, 20.4, 25.9],
    'Ingiliz tili':[18.7, 18.5, 20.1, 17.3, 15.9, 13.7, 14.2, 11.2]
} #lugat hosil qilib oldik
```

```python
ball = pd.DataFrame(data) # ball nomli lugatni jadval korinishiga keltirib oldik
```

```python
ball
```

```python
ball.index = [8, 7, 6, 5, 4, 3, 2, 1]
ball # index ketma ketligini 1 dan 8 gacha qilib oldik
```

```python
tartib = [1, 2, 3, 4, 5, 6, 7, 8]
tartib.sort()
tartib # 1 dan 8 gacha bolgan lugat hosil qilib oldik va uni sort orqali tartiblab chiqdik
```

```python
ball = ball.reindex(tartib) # jadvalimizvagi indexlarimiz reindex orqali qata indekslandi
```

```python
ball
```

```python
ball.reindex([3,5,8]) # jadvaldan ozimizga kerak bolgan malumotni chiqarib oldik
```

```python
ball.reindex(index=range(1,11))
# reindex va range orqali 1 van 11 gacha bogan malumot qoshdik bizda 9 va 10 malumot mavjud bolmaganligi uchun NAN berdi
```

```python
ball.reindex(index=range(1,11), method = 'nearest')
# bu ozgincha ozgarish qildik method va nearest orqali biz NAN bolgan qimatni oldidagi malumotlarni kochirib oldim
```

```python
ball.reindex(index=range(1,11), method='ffill')
# bu ham tepadagiga oxshidi faqat bunda 'ffill' dan fodalanildi bu uzidan tepada bolgan mulotni olib hammasini bir xil qiladi
```

```python
ball.reindex(columns=['Talaba', 'Informatika', 'Ingiliz tili'])
# reindex va columns orqali ozimizga kerakli bolgan jadvalni olib keraksizini chiqarib ubordik
```

```python
ball.sample() # bu funksia orqali biz ga tasodifi qimatni olib berdi
```
