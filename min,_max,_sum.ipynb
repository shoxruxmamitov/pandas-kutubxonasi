```python
import pandas as pd
```

```python
data = {
    'Aholi':[123, 217, 318, 756, 761, 813, 522],
    'Madoni':[1290, 2151, 3127, 2999, 31871, 5312, 3287]
}
```

```python
Jadval = pd.DataFrame(data)
Jadval
```

```python
Jadval.index = (1, 2, 3, 4, 5, 6, 7)
Jadval
```

```python
Jadval.index.name = '№'
Jadval
```

```python
Jadval.sum() # jami sonni chiqarib oldik
```

```python
Jadval.sum(axis='columns') #ustun boicha jami sonni chiqarib oldik
```

```python
Jadval.sum(axis=1) #buni ham ustun boicha jami sonni chiqarib oldik
```

```python
Jadval.mean(axis=0)
# qator boicha ortancha qimatni topib oldik
```
