```python
warr = pd.DataFrame(np.random.randn(10, 5))
warr
```

```python
warr.shape # qator va ustunini korib oldik
```

```python
warr.size # qimatini korib oldik
```

```python
warr.sort_index() # tartiblab oldik
```

```python
warr.sort_index(axis=0) # qator boicha tartiblimiz
```

```python
warr.sort_index(axis=1) # ustun boicha tartiblimiz
```

```python
warr.sort_index(axis=1, ascending=False) # ustun boicha tartiblidi
```

```python
warr.sort_index(axis=0, inplace=True) # qator boicha tartiblidi
```

```python
warr
```

```python
warr.sort_index(ascending=False)
warr # # teskari osish tartibida tahlab beradi
```

```python
warr.sort_index(axis=1, ascending=False) # qator boicha kamaish tartibidi tartiblidi
```

```python
warr.sort_values(by=2) # 2 nomli ustunni osish tartiblab tartiblab chiqdik chiqdik
```

```python
warr.sort_values(by=2, ascending=False)
warr # 2 nomli ustunnu kamaish tartibida tartiblab chiqdik
```

```python
warr.sort_values(by=[1, 4], ascending=False)
# 1va 4 chi ustunni kamaish tartibida tartibla chiqdik
```
