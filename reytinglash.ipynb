```python
data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv", index_col=0)
data # dataframe hosil qilib oldik
```

```python
data.sort_values(by='sepal_width') # sepal_width ustunni tartiblab oldik
```

```python
data['form'] = data['sepal_width'].rank()
# form nomli ustun qoshib oldik va sepal_width ustun bn bogliq qilib qodik
```

```python
data.sort_values('sepal_width') #sepal width ni tartibladik va form avtomatski sepal widthga qarab tartiblandi chunki ikklasini tepada bogladik
```

```python
data['form'] = data['sepal_width'].rank(method = 'min')
# form ustunini sepal_width ga boglab qodik va minimum boicha tartibladik
```

```python
data.sort_values('sepal_width') # sepal_widthni artibladik va form avtomatski tartiblandi
```

```python
data.sort_values('petal_length')
# petal_length ni tartibladik va form avtomatski tatriblandi
```
