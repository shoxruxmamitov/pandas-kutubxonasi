```python
import pandas as pd
import numpy as np
```

```python
t1 = pd.Series([40, -30, 2.7, 91, 35, 71, 98.2, 32.1, -83], index = ['x', 'm', 'n', 'p', 'o', 'u', 'k', 'l', 'j'])
t2 = pd.Series([31, -19, 1.5, 82, 21, 39, 71.9], index = ['x', 'm', 'n', 'p', 'o', 'u', 'k'])

# t1 va t2 Series hosil qilib oldik index qatorlarini Series ichida hosil qilib oldik
```

```python
t1+t2 # t1 vat2 elementarni qoshidik sonlar bir biriga tugri kelmaganligi uchun nan berdi
```

```python
set1 = pd.DataFrame(np.arange(9.).reshape(3, 3), columns=list('abc'), index=['Kartoshka', 'Sabzi', 'Guruch'])
set2 = pd.DataFrame(np.arange(12.).reshape(4, 3), columns=list('klm'), index=['Go\'sht', 'Guruch', 'Tuz', 'Murch'])

# set1 set2 dataframe hosil qilib unga 3 ga 4 qatorli sonlarni va indexga abc va klm qatoriga nom berib qimatini hosil qilib oldik
```

```python
set1
```

```python
set2
```

```python
set1+set2 # set1 av set2 ni qoshib oldik sonlar bir biriga tugri kelmaganlii uchun NAN berdi
```

```python
set1.add(set2) # add orqali qoshish kodini bajardik
```

```python
set1.add(set2, fill_value=0) # oxiri jadva korinishida qoshib keladi
```

```python
set2.sub(set1) # bu airish funksiasi
```

```python
set1.sub(set2, fill_value=0) # bu ham jadval oxiridan airb keladi
```

```python
set2.rsub(set1, fill_value=0) # ornini almashtirib airish
```

```python
set1.div(set2) # bo'lish funksiasi
```

```python
set2.div(set1)
```

```python
set2.rdiv(set1) # bu bolish funksiasi
```
